Los principales problemas que tenemos con esos sistemas son los siguientes:

- En casi todos ellos el código está puesto como `PLATINO`, por lo que **no se pueda usar como codigo identificativo** del sistema.

- Los **nombres no son muy descriptivos**.

- La **funcionalidad está austente** en todos los casos escepto en uno.

- Algunos de los sistema se parecen demasiado, lo que lleva a pensar que quizá **puedan considerarse como un  único sistema**.

## Cambios sugeridos

Los cambios que yo haría a estas entradas de SS.I. de la ICAVI son los siguientes:

1) La aplicación con el nombre `AUTOCONSTRUCCIÓN` y código `APPSACCESS`, cambiaría el nombre
   a `Autoconstrucción` y el código a `AUTOCONSTRUCCION`. Podemos guardar en el campo de comentarios el código antguo, ya que parece un indicatio de que se trata de una aplicación en MS Access. Habría que ferificarlo.

2) El sistema `DISCIPLINA PÚBLICA Y PRIVADA` no puede tener como código `PLATEA` (Esto sucede con el resto de los sistemas listados), se sugiere cambiar `EXP_DISCIPLINARIOS`.

3) Los tres sistemas siguientes, a falta de datos en el campo de finalidad, se sugiere reunificarlos en un único sistema. Las diferencias de tratamiento en lo que respecta a promoción pública, privada, descalificaciones, etc. se pueden analizar en el siguiente nivel de activos de datos. Se sugiere cógigo `CALIFICACION_VIVIENDAS` y como nombre `Gestión de calificación de viviendas`.

4) Para `RECURSOS ADMINISTRATIVOS`, se sugiere el código `REC_ADM_VIVIENDAS`.

5) Para `SUBVENCION MEDIANTE CONCURRENCIA COMPETITIVA  CONSTRUCCION DE VIVIENDAS`, se sugieren dos posibles soluciones:

   - Incluirlo en un sistema general de subvenciones (aunque esté compuesto por varias aplicaciones) y dejar para la fase  de análisis de activos de datos el realizar las distinciones que correspondan entre todos los tipos de subvenciones.

   - Darlo de alta como un S.I. propio. Se sugiere como código, en ese caso, `SUB_CONSTRUCCION_VIVIENDAS`.

6) Para `REINTEGROS`, cambiar el nombre a `Gestión de reintegros` y el código a `REINTEGROS`

7) Para `CONTRATOS DE SERVICIO EN MATERIA DE VIVIENDA`, se sugiere cambiar el código a `CONT_SERVICIOS_VIV` y la descripción a `Gestión de contratos de servicio en materia de vivienda`.

8) Para `SUBVENCIONES PARA LA ADQUISICIÓN DE SUELO` la propuesta es similar a la del punto 5, o bien integrarlo como otro tipo de subvenciones en un S.I. único de subvenciones, o si se opta por dejarlo como un sistema diferencias, cambiar el nombre por `Subvenciones para la adquisición de suelo`, y el código a `SUB_ADQUISICION_SUELO`.

9) Para este último caso, ver si se puede incluir como una operación más de los tres sistemas unificados en el punto 3, o bien se opta por dejarlo como sistema aparte. Yo recomiendo la primera, a no ser que haya razones justificadas para la división. Si se optara por esta segunda opción , sugiero como nombre `Cambio de uso vivienda` y como código `CAMBIO_USO_VIVIENDAS`.
